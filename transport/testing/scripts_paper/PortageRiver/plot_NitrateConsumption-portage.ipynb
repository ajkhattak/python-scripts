```python
# Get number of cells per reach in the coarse scale simulation
C = Concentration[0] #np.arange(1,len(Concentration[0]), 1)
sim_cells_d = np.loadtxt(os.path.join(infile,'reach_cells/vector_cells-%sm_min%sm.dat'%(dx,dx_min)),usecols=(0,2,3),delimiter=',')
sim_cells = np.array(sim_cells_d[:,0],dtype=int)
sim_bccells = np.array(sim_cells_d[:,1],dtype=int)
reach_lengths = np.array(sim_cells_d[:,2],dtype=int)
#print (sim_bccells)
def get_cres_cells(sim_cells_d):
    csum = np.cumsum(sim_cells)
    for c in csum:
        yield c
```

```python
# Read discharge
def get_mapped_Q():
    Q_file = h5py.File(os.path.join(infile,'discharge/discharge-%sm_min%sm.h5'%(dx,dx_min)),'r')
    discharge = np.concatenate(Q_file['surface-mass_flux.face.0']['0'][:])

    ptr = get_cres_cells(sim_cells)
    Q1 = []
    xprev = 0
    bc_count = 0
    for i,p in enumerate(ptr):
        if sim_bccells[i] == 0:
            Q1.append(discharge[xprev+bc_count:p+bc_count])
            bc_count = bc_count + 1
            xprev = p 
        else:
            Q1.append(discharge[xprev+bc_count:p+bc_count])
            xprev = p

    Q1 =  np.concatenate(Q1)
    print (len(Q1))
    return Q1
    
coarseQ = get_mapped_Q()
```

```python
#Read source
Src_file = h5py.File(os.path.join(infile,'implied_source/implied_watersourceQW-%sm_min%smA.h5'%(dx,dx_min)),'r')
coarseQW = np.concatenate(Src_file['surface-water_mass_source.cell.0']['0'][:])
print (len(coarseQW),coarseQW[:5])
```

```python
def get_mapped_data(coarseC, coarseQ, coarseQW,r_lor, r_hor):
    p_mesh_res = get_hres_cells(cellsmerged_d)
    p_mesh_sim = get_cres_cells(sim_cells_d)
    xprev1, xprev2 = 0, 0
    fineC, fineQ, fineQW = [], [], []
    fineC_NR = []
    coarseC_NR,coarseC_A = [], []
    N_source = []
    Qin_v, Qout_v, Qgw_v = [], [], []
    meshx = []
    meshy = []
    bc_count = 0
    c_x = 0
    count = 0
    for it1, it2 in zip(p_mesh_res,p_mesh_sim):
        l1 = len(coarseC[xprev2:it2])
        l1 = 1 if l1 ==0 else l1
        l2 = int((it1 - xprev1)/(it2 -xprev2))
        C_temp = np.zeros((l1,l2))
        Q_temp = np.zeros((l1,l2))
        QW_temp = np.zeros((l1,l2))
        for i in range(l1):
            C_temp[i] = coarseC[xprev2:it2][i]
            Q_temp[i] = coarseQ[xprev2:it2][i]
            QW_temp[i] = coarseQW[xprev2:it2][i]
            
        c_x = c_x + it2 -xprev2
        
        for c1,c2,c3 in zip(C_temp,Q_temp,QW_temp):
            fineC.append(c1)
            fineQ.append(c2)
            fineQW.append(c3)
          
        fineC1 = [c for x in fineC for c in x]
        
        r = (it1  - len(fineC1))
        for i1 in range(r):
            fineC.append([fineC[-1][0]])
            fineQ.append([fineQ[-1][0]])
            fineQW.append([fineQW[-1][0]])
        
        
        #Nitrate Consumption
        A, B, C = False, False, False
        if sim_bccells[bc_count] == 0:
            xx = np.sum([c4 * r_lor for c4 in coarseQW[xprev2:it2]])
            #*src_load_rate = 0*Q_temp[-1][0] * C_temp[-1][0] + QW_temp[-1][0] * r_lor * (it2-xprev2)
            src_load_rate = np.sum([c4 * r_lor for c4 in coarseQW[xprev2:it2]])
            src_load = r_lor * (it2-xprev2)
            A = True
        elif (QW_temp[-1][0] > 0): #gaining reach
            #*src_load_rate = Q_temp[-1][0] * C_temp[-1][0] + QW_temp[-1][0] * r_hor * (it2-xprev2)
            src_load_rate = Q_temp[-1][0] * C_temp[-1][0] + np.sum([c4 * r_hor for c4 in coarseQW[xprev2:it2]])
            src_load =  C_temp[-1][0] +  r_hor * (it2-xprev2)
            B= True
        else:
            #*src_load_rate = Q_temp[-1][0] * C_temp[-1][0] 
            src_load_rate = coarseQ[xprev2:it2][-1] * C_temp[-1][0]
            src_load = C_temp[-1][0]
            print ('AA: ', Q_temp[-1][0], coarseQ[xprev2:it2][-1])
            C = True
        
        qout_v = (Q_temp[0][0] * C_temp[0][0])
        cout_v = (Q_temp[0][0] * C_temp[0][0])
        if (QW_temp[-1][0] < 0): #losing reach
            #*qout_v = qout_v + QW_temp[-1][0] *50 * (it2-xprev2)
            qout_v = qout_v + np.sum([c4 * r_hor for c4,c5 in zip(coarseQW[xprev2:it2],coarseC[xprev2:it2])])
            cout_v = cout_v + 50 * (it2-xprev2)
            #print ('A:', count, QW_temp[-1][0], r_hor, coarseC[xprev2:it2])
        
        
        nr_v = 100 * ( src_load_rate - qout_v)/ (src_load_rate * reach_lengths[count]/1000.)  #per km
        nr_v = max(nr_v, 0.001) # this is only because of the numerical dispersion
        
        nr_a = 1000 * (src_load_rate - qout_v)*12*1e-6 #/ reach_lengths[count] #moles to gram
        
        fineC_NR.append([nr_v,]*(it1-xprev1))
        coarseC_NR.append([qout_v])
        coarseC_A.append([nr_a])
        N_source.append([src_load_rate])
        
        if sim_bccells[bc_count] == 0:
            Qin_v.append([Q_temp[-1][0]*0])
        else:
            Qin_v.append([Q_temp[-1][0]])
        Qout_v.append([Q_temp[0][0]])
        Qgw_v.append([QW_temp[-1][0]])
        
        count = count + 1
        xprev1 = it1
        xprev2 = it2
        bc_count = bc_count + 1
    fineC = [c for x in fineC for c in x]
    fineQ = [c for x in fineQ for c in x]
    fineC_NR = [c for x in fineC_NR for c in x]
    coarseC_NR = [c for x in coarseC_NR for c in x]
    coarseC_A = [c for x in coarseC_A for c in x]
    N_source = [c for x in N_source for c in x]
    Qin_v = [c for x in Qin_v for c in x]
    Qout_v = [c for x in Qout_v for c in x]
    Qgw_v = [c for x in Qgw_v for c in x]
    #for i,q in enumerate(Qin_v):
    #    print ('Qin = ',i, q)
    return fineC, fineQ, fineC_NR, coarseC_NR,N_source, Qin_v, Qout_v, Qgw_v, coarseC_A
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

```
