```python
path = '/Users/ajc/FUSE2/simulations/tests/ats-demos/13_reactive_transport/01_surface'
```

```python
# Read PFOTRAN standalone data
pflotran_file = os.path.join(path, 'data/pflotran_out/denitrification_reach-001.tec')
plfotran_data = dict()
#plfotran_data['time'] = np.loadtxt(pflotran_file,usecols=3,skiprows=3) #time in hours
plfotran_data['X'] = np.loadtxt(pflotran_file,usecols=0,skiprows=3)  # Tracer 1
plfotran_data['DOM1'] = np.loadtxt(pflotran_file,usecols=3,skiprows=3)  # Tracer 1
plfotran_data['O2(aq)'] = np.loadtxt(pflotran_file,usecols=4,skiprows=3)  # Tracer 1
plfotran_data['NO3-'] = np.loadtxt(pflotran_file,usecols=5,skiprows=3)  # Tracer 1
print (plfotran_data['X'])
```

```python
F= ['srun4d']
Files = [os.path.join(path,f,'visdump_surface_data.h5') for f in F]
vars = ['surface-total_component_concentration', 'surface-total_component_concentration','surface-total_component_concentration']
Suffix = ['.cell.DOM1 conc', '.cell.NO3- conc', '.cell.O2(aq) conc']
```

```python
#%matplotlib tk
fig, axs = plt.subplots(1,1, figsize=(6,4), facecolor='w', edgecolor='k',gridspec_kw={'width_ratios':[1]})
fig.subplots_adjust(hspace =.02, wspace=1.02)
plt.tight_layout()
#axs = axs.ravel()

Colors = ['r','k','m']
F_L = ['DOM1', 'NO3', 'O2']
for i, v in enumerate(vars):

    plt.subplot(111)
    for j,f in enumerate(Files):
        dat = readdata.GetSurfVarFromVis_All(f,v,suffix=Suffix[i])[v]
        #print (dat)
        data = np.array(dat[-1])
        #if i ==0:
        #    print (data[:10], len(data))
        #break
        #Time = np.array([240*i/3600. for i in range(len(data))])
        plt.semilogy(plfotran_data['X'],1e-3*data,color=Colors[i],label=F_L[i],linewidth=2)
    
    plt.ylabel(vars[i] + ' [M]') # Molarity = Moles/liter
    plt.xlabel('Length [m]')
    
plt.semilogy(plfotran_data['X'],plfotran_data['DOM1'],'k',linestyle='--',label='Pflotran DOM1')
plt.semilogy(plfotran_data['X'], plfotran_data['NO3-'],'g',linestyle='--',label='Pflotran NO3-')
plt.semilogy(plfotran_data['X'], plfotran_data['O2(aq)'],'c',linestyle='--',label='Pflotran O2(aq)')
plt.legend(loc='lower left',fontsize=10)
#plt.title('Location = 1000 m')
plt.ylim(1e-6,1e-3)
#plt.xlim(0,50)
#print (plfotran_data['DOM1'])
plt.show()
```
